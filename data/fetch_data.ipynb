##### Get historic price data for each category of items

**Find items in relevant categories**
Don't need to fetch prices for every item in the game, only those I'm interested in training on which is those who can be traded on the Trading Post.
Possible categories: Armor, Back, Bag, Consumable, Container, CraftingMaterial, Gathering,MiniPet, PowerCore, Trinket, UpgradeComponent, Weapon

Should also filter out items flagged as being accountbound or soulbound
Flags: AccountBound, NoSell, SoulbindOnAcquire

```python
filter_condition = lambda flags: all(item not in flags for item in ["SoulbindOnAcquire", "AccountBound", "NoSell"])
filtered_df = item_info[item_info['flags'].apply(filter_condition)]

categories = ["Armor", "Back", "Bag", "Consumable", "Container", 
              "CraftingMaterial", "Gathering", "MiniPet", "PowerCore", 
              "Trinket", "UpgradeComponent", "Weapon"]

filtered_items = filtered_df[filtered_df['type'].isin(categories)]
```

## Split items into categories

```python
armor = filtered_items[filtered_items["type"] == "Armor"]["id"].tolist()
back = filtered_items[filtered_items["type"] == "Back"]["id"].tolist()
bag = filtered_items[filtered_items["type"] == "Bag"]["id"].tolist()
consumable = filtered_items[filtered_items["type"] == "Consumable"]["id"].tolist()
container = filtered_items[filtered_items["type"] == "Container"]["id"].tolist()
crafting_material = filtered_items[filtered_items["type"] == "CraftingMaterial"]["id"].tolist()
gathering = filtered_items[filtered_items["type"] == "Gathering"]["id"].tolist()
mini_pet = filtered_items[filtered_items["type"] == "MiniPet"]["id"].tolist()
power_core = filtered_items[filtered_items["type"] == "PowerCore"]["id"].tolist()
trinket = filtered_items[filtered_items["type"] == "Trinket"]["id"].tolist()
upgrade_component = filtered_items[filtered_items["type"] == "UpgradeComponent"]["id"].tolist()
weapon = filtered_items[filtered_items["type"] == "Weapon"]["id"].tolist()
```
